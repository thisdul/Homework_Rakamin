## Load Data

```python
# import library pandas
import pandas as pd
```

```python
# upload dataset
dataset = pd.read_csv("online_retail_II.csv")

df = pd.DataFrame(dataset)
# df.head(10)
df
```

```python
df.dtypes
```

## Section 1

### Create New Feature: Year

```python
# Make new column for years

# change the data type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
# df["InvoiceDate"] = df["InvoiceDate"].astype('datetime64[ns]')

year = df['InvoiceDate'].dt.year

df['Year'] = year

df
```

```python
df.dtypes
```

### Filtering Data

```python
# df_dataset[ (df_dataset['province'] == 'Jawa Timur') & (df_dataset['age'] > 40) ]
# df_dataset[df_dataset['city'] != 'Malang']

# Quantity minimal 1 (tidak boleh 0 dan minus)
# Kolom Invoice tidak mengandung huruf ‘C’ karena hal tersebut menandakan pelanggan tidak menyelesaikan belanjanya atau melakukan pembatalan

df_dataset = df[(df['Quantity'] >= 1) & (df['Invoice'] != 'C')]

df_dataset
```

### Create New Feature: Revenue

```python
# buat kolom baru
# Buat kolom baru bernama Revenue dengan nilai Quantity dikali dengan Price
# Menambahkan kolom baru
# df['Nilai Akhir'] = df['Tugas']*0.3 + df['UTS']*0.3 + df['UAS']*0.4
# df
df_dataset['Revenue'] = df_dataset['Quantity'] * df_dataset['Price']
df_dataset


```

### Average of Revenue per Year

```python
# Hitung rata-rata Revenue per tahun. Lalu buatlah visualisasinya.
data_vis = df_dataset.groupby('Year')['Revenue'].mean().reset_index()
data_vis
```
